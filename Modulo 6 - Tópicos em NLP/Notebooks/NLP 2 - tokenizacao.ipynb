```python
import nltk
```

```python
with open("texto_exemplo.txt") as f:
    texto = f.read()
texto
```

```python
texto.split(' ')[:40]
```

```python
nltk.download('punkt')
```

Tokenização por sentenças

```python
sentencas = nltk.sent_tokenize(texto)
sentencas
```

Tokenização por palavras

```python
palavras = nltk.word_tokenize(texto)
palavras
```

```python
palavras = nltk.word_tokenize(sentencas[1])
palavras
```
