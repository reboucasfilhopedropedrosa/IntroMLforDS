## Métricas de avaliação

#### Imports

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

#### 1. Métricas de regressão

```python
data = load_breast_cancer()
X = data.data
y = data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
specificity = tn / (tn + fp)
```

```python
print(f"Acurácia: {accuracy}")
print(f"Precisão: {precision}")
print(f"Recall: {recall}")
print(f"F1-Score: {f1}")
print(f"Especificidade: {specificity}")
```

1. **Acurácia (97.08%)**: 
   - Indica a proporção geral de previsões corretas (positivas e negativas).
   - Uma acurácia de 97.08% significa que o modelo classifica corretamente 97.08% das instâncias.

2. **Precisão (96.40%)**:
   - Reflete a proporção de identificações positivas que são realmente corretas.
   - Com precisão de 96.40%, quando o modelo prevê uma instância como positiva, há 96.40% de chance de estar correto.

3. **Recall (99.07%)**:
   - Mede a capacidade do modelo de identificar todas as instâncias positivas.
   - Um recall de 99.07% indica que o modelo captura 99.07% de todas as instâncias positivas reais.

4. **F1-Score (97.72%)**:
   - É a média harmônica de precisão e recall.
   - Um F1-Score de 97.72% mostra um equilíbrio entre não perder positivos (alto recall) e estar correto nas afirmações positivas (alta precisão).

5. **Especificidade (93.65%)**:
   - Indica a proporção de verdadeiros negativos corretamente identificados.
   - Com uma especificidade de 93.65%, o modelo é eficaz em identificar instâncias negativas, mas não tanto quanto em identificar positivas.
